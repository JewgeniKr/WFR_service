```python
import cv2
from yolo import detect as yolo_detect
from ultralytics import YOLO
# загружаем модель
model = yolo_detect.load('runs/detect/train15/weights/last.pt')
img_path = '0_waybill_date_conf_0.96.jpg'
img = cv2.imread(img_path)

results = model([img])
bboxes = results.xyxyn[0].cpu().numpy() # координаты bbox'ов в нормализованных значениях
class_ids = bboxes[:, -1].astype(int)   # класс каждого объекта
confidences = bboxes[:, -2]
bbox_coords = bboxes[:, :4]              # нормированные координаты x1,y1,x2,y2

for i in range(len(bbox_coords)):
    class_id = class_ids[i]
    confidence = confidences[i]
    
    # денормализованные координаты
    h, w, _ = img.shape
    x1, y1, x2, y2 = map(int, bbox_coords[i]*[w,h,w,h])
    
    color = (0, 255, 0) if class_id == 0 else (0, 0, 255) # зеленый для первого класса, красный для остальных
    thickness = 2
    
    # рисуем рамку
    cv2.rectangle(img, (x1, y1), (x2, y2), color, thickness)
    
    # добавляем метку класса и уверенность
    label = f'{class_id}: {confidence:.2f}'
    cv2.putText(img, label, (x1, y1-10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, color, thickness=2)

cv2.imshow("Detected Objects", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
from ultralytics import YOLO
import cv2
import numpy as np
import os

# Загрузка модели YOLOv8
model = YOLO('runs/detect/train15/weights/best.pt')

# Список цветов для различных классов
colors = [
    (255, 0, 0), (255, 20, 147)
]

# Функция для обработки изображения
def process_image(image_path):
    # Загрузка изображения
    image = cv2.imread(image_path)
    results = model(image)[0]
    
    # Получение оригинального изображения и результатов
    image = results.orig_img
    classes_names = results.names
    classes = results.boxes.cls.cpu().numpy()
    boxes = results.boxes.xyxy.cpu().numpy().astype(np.int32)

    # Подготовка словаря для группировки результатов по классам
    grouped_objects = {}

    # Рисование рамок и группировка результатов
    for class_id, box in zip(classes, boxes):
        class_name = classes_names[int(class_id)]
        color = colors[int(class_id) % len(colors)]  # Выбор цвета для класса
        if class_name not in grouped_objects:
            grouped_objects[class_name] = []
        grouped_objects[class_name].append(box)

        # Рисование рамок на изображении
        x1, y1, x2, y2 = box
        cv2.rectangle(image, (x1, y1), (x2, y2), color, 1)
        cv2.putText(image, class_name, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 1)

    # Сохранение измененного изображения
    new_image_path = os.path.splitext(image_path)[0] + '_yolo' + os.path.splitext(image_path)[1]
    cv2.imwrite(new_image_path, image)

    print(f"Processed {image_path}:")
    print(f"Saved bounding-box image to {new_image_path}")

process_image('0_waybill_date_conf_0.96.jpg')
```

```python
from ultralytics import YOLO
import cv2
import numpy as np
import os

# Загрузка модели YOLOv8
model = YOLO('runs/detect/train15/weights/best.pt')

# Функция для обработки изображения
def process_image(image_path):
    # Загрузка изображения
    image = cv2.imread(image_path)
    results = model(image)[0]
    
    # Получаем оригинальное изображение и результаты детекции
    orig_image = results.orig_img.copy()  # Создаем копию исходного изображения
    classes_names = results.names
    classes = results.boxes.cls.cpu().numpy()
    boxes = results.boxes.xyxy.cpu().numpy().astype(np.int32)

    # Подготовка папки для сохранения объектов
    output_dir = 'output'
    os.makedirs(output_dir, exist_ok=True)

    # Обработка каждого обнаруженного объекта
    for idx, (class_id, box) in enumerate(zip(classes, boxes)):
        class_name = classes_names[int(class_id)]
        
        # Вырезаем область вокруг объекта
        x1, y1, x2, y2 = box
        cropped_object = orig_image[y1:y2, x1:x2]
        
        # Генерируем уникальное имя файла
        filename = f"{os.path.basename(os.path.splitext(image_path)[0])}_{idx+1}.jpg"
        save_path = os.path.join(output_dir, filename)
        
        # Сохраняем вырезанный объект
        cv2.imwrite(save_path, cropped_object)
        print(f"Сохранено изображение класса '{class_name}' в {save_path}")

# Запускаем обработку изображения
process_image('0_waybill_date_conf_0.96.jpg')
```
