```python
from ultralytics import YOLO
import cv2

# Загрузка модели YOLO
model = YOLO("runs/detect/train/weights/best.pt")  # или другая версия YOLOv8

# Загрузка изображения
image = cv2.imread("test5.png")

# Предсказание объектов
results = model(image, imgsz=640, iou=0.4, conf=0.8)

# Получение bounding box'ов и классов
boxes = results[0].boxes.xyxy  # Координаты bbox в формате [x1, y1, x2, y2]
classes = results[0].boxes.cls  # Классы объектов
confidences = results[0].boxes.conf  # Уверенность модели
```

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

train_results = pd.read_csv('C:/Users/e.krukov/Desktop/ML_notebooks/waybill_page1_full_labled/runs/detect/train/results.csv')

plt.figure(figsize=(10, 6))

plt.plot(train_results['epoch'], train_results['train/cls_loss'], label='Train Loss', color="blue")
plt.plot(train_results['epoch'], train_results['val/cls_loss'], label='Validation Loss', color="orange")

plt.title("Значения функции потерь на ренировочном и валидационном наборах данных")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 6))

plt.plot(train_results['epoch'], train_results['metrics/mAP50(B)'], label='mAP50', color="green")
plt.plot(train_results['epoch'], train_results['metrics/mAP50-95(B)'], label='mAP50-95', color="purple")

plt.title("Значения точности")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import os

# Создаем папку для сохранения объектов
output_dir = "detected_objects"
os.makedirs(output_dir, exist_ok=True)

# Перебираем все обнаруженные объекты
for i, (box, cls, conf) in enumerate(zip(boxes, classes, confidences)):
    x1, y1, x2, y2 = map(int, box)  # Преобразуем координаты в целые числа
    
    # Обрезаем объект из изображения
    object_img = image[y1:y2, x1:x2]
    
    # Получаем имя класса (если доступно)
    class_name = model.names[int(cls)]  # для YOLOv8
    
    # Сохраняем обрезанное изображение
    output_path = f"{output_dir}/object_{i}_class_{class_name}_conf_{conf:.2f}.jpg"
    cv2.imwrite(output_path, object_img)

print(f"Сохранено {len(boxes)} объектов в папку {output_dir}")
```
